```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.inspection import permutation_importance
from sklearn.svm import SVC 
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from catboost import CatBoostClassifier
import lightgbm as lgb

scaler = StandardScaler()
rs = RobustScaler()
mms = MinMaxScaler()

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV

import warnings
warnings.filterwarnings("ignore")
```

```python
faults = ['Other_Faults', 'Bumps', 'K_Scatch', 'Z_Scratch', 'Pastry', 'Stains', 'Dirtiness']
```

```python
data_wfe = pd.read_csv('datasets/data_wfe.csv') 
data_wfe.head() 
```

```python
data = pd.read_csv('datasets/data_fe.csv') 
data.head() 
```

```python
data.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
def clsf(data, data_wfe,
         param_grid, 
         model, 
         cv=5,
         random_state=40):
    
    k = 1 
    while k < 3:
        
        # train test split
        X_train, X_test, y_train, y_test = train_test_split(
            data.drop(['Defects'], axis=1), 
            data['Defects'], test_size=0.2, 
            random_state=42, 
            stratify=data['Defects']
        )
        # standard scaler
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
        # min max scaler
        X_train_minmax = mms.fit_transform(X_train)
        X_test_minmax = mms.transform(X_test)
        # robust scaler
        X_train_rs = rs.fit_transform(X_train)
        X_test_rs = rs.transform(X_test)
        
        search_scl = RandomizedSearchCV(model, param_grid, cv=cv, 
                                        random_state=random_state)
        
        search_scl.fit(X_train_scaled, y_train)
        
        search_mnx = RandomizedSearchCV(model, param_grid, cv=cv, 
                                        random_state=random_state)
        
        search_mnx.fit(X_train_minmax, y_train)

        search_rbs = RandomizedSearchCV(model, param_grid, cv=cv, 
                                        random_state=random_state)
        
        search_rbs.fit(X_train_rs, y_train)
     
        # best score betwwen standard, min max and robust
        best_score = max([search_scl.best_score_, search_mnx.best_score_, 
                          search_rbs.best_score_]) 

        if best_score == search_scl.best_score_:
            
            test_predictions = search_scl.best_estimator_.predict(X_test_scaled)
            report = classification_report(y_test, test_predictions, 
                                           target_names=faults)

            print(f"Best CV score: {search_scl.best_score_:.3f}(Standard Scaler)",
                  f"\nBest parameters:" 
                  f"{search_scl.best_params_ if bool(param_grid) else '-'}\n\n",
                  report) 
            
            search = search_scl
            X = X_train_scaled

        elif best_score == search_mnx.best_score_:
            
            test_predictions = search_mnx.best_estimator_.predict(X_test_minmax)
            report = classification_report(y_test,test_predictions, 
                                           target_names=faults)

            print(f"Best CV score: {search_mnx.best_score_:.3f}(Min-max)",
                  f"\nBest parameters:" 
                  f"{search_mnx.best_params_ if bool(param_grid) else '-'}\n",
                  report) 
            
            search = search_mnx
            X = X_train_minmax
            
        elif best_score == search_rbs.best_score_:
            
            test_predictions = search_rbs.best_estimator_.predict(X_test_rs)
            report = classification_report(y_test,test_predictions, 
                                           target_names=faults)
            
            
            print(f"Best CV score: {search_rbs.best_score_:.3f}(Robust Scaler)",
                  f"\nBest parameters:" 
                  f"{search_rbs.best_params_ if bool(param_grid) else '-'}\n",
                  report) 
            
            search = search_rbs
            X = X_train_rs
            
        results = permutation_importance(search, X, y_train, 
                                         scoring='neg_mean_squared_error')
        
        importance = results.importances_mean
        
        feature_importance_score = []
        for i, v in zip(data.columns, importance):
            feature_importance_score.append(v)
            
        fig = plt.figure(figsize=(13,8))
        plt.barh(X_train.columns, feature_importance_score, color='midnightblue')
        plt.title('Feature importance')
        plt.show()  
            
        print('Без добавления новых признаков') if k == 1 else print('')
        k = k + 1
        # Набор данных без генерации признаков для второго цикла
        data = data_wfe.copy()
```

### Parameters grid

##### KNN

```python
param_grid_knn = {
    'n_neighbors': range(1, 50),
    'p': range(1, 10),
    'metric': ['euclidean', 'manhattan', 'chebyshev']
}
```

##### Naive Bayes

```python
param_grid_nb = {}
```

##### LGBM

```python
param_grid_lgbm = {
    'learning_rate': [0.005, 0.01],
    'n_estimators': [8,16,24,50,100],
    'num_leaves': [6,8,12,16], 
    'boosting_type' : ['gbdt', 'dart'], 
    'objective' : ['binary'],
    'max_bin':[255, 510], 
    'random_state' : [500],
    'colsample_bytree' : [0.64, 0.65, 0.66],
    'subsample' : [0.7,0.75],
    'reg_alpha' : [1,1.2],
    'reg_lambda' : [1,1.2,1.4],
    }
```

#### LGBM
